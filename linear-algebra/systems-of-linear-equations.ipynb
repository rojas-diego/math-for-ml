# Systems of linear equations

In mathematics, a system of linear equations (or linear system) is a collection of one or more linear equations involving the same variables.

For example, the following linear system.

$\begin{align}
    4x_1 + 4x_2 &= 5 \\
    2x_1 + 4x_2 &= 1
\end{align}$

Has a single solution $(x_1, x_2) = (1, \frac{1}{4})$ .

That we can obtain the following way.

$\begin{align}
    4x_1 + 4x_2 &= 5 \\
    2x_1 + 4x_2 &= 1
\end{align}$

> ðŸ’œ In general, for a real-valued system of linear equations we obtain either **no**, **exactly one**, or **infinitely many solutions**.

In other words, we can obtain a solution set that is a plane, a line, a point or empty.

Let's take a look at another linear system.

A company produces products $N_1, \dotsc, N_n$ for which resources $R_1, \dotsc, R_j$ are required. To produce a unit of product $N_j$, $a_{ij}$ units of resource $R_i$ are needed. The objective is to find an optimal production plan, i.e., a plan of how many units $x_j$ of product $N_j$ should be produced if a total of $b_i$ units of resource $R_i$ are available and (ideally) no resources are left over.

If we produce $x_1, \dotsc, x_n$ units of the corresponding products, we need this many units of resource $R_i$.

$
a_{i1} x_1 + \dotsc + a_{in}x_n
$

An optimal production plan $(x_1, \dotsc, x_n) \in \R^n$, therefore, has to satisfy the following system of equations:

$\begin{align}
a_{11} x_1 + \cdots +& a_{1n} xn = b_1 \\
                   \vdots & \\
a_{m1} x_1 + \cdots +& a_{mn} x_n = b_m
\end{align}$


where $a_{ij} \in \R$ and $b_i \in \R$.

This linear system is often written as the following compact notation:

$\begin{bmatrix}
    a_{11} \\
    \vdots \\
    a_{m1}
\end{bmatrix} x_1 + \begin{bmatrix}
    a_{12} \\
    \vdots \\
    a_{m2}
\end{bmatrix} x_1 + \cdots + \begin{bmatrix}
    a_{1n} \\
    \vdots \\
    a_{mn}
\end{bmatrix} x_n = \begin{bmatrix}
    b_{1} \\
    \vdots \\
    b_{m}
\end{bmatrix} \iff \begin{bmatrix}
    a_{11} & \cdots & a_{1n} \\
    \vdots & & \vdots      \\
    a_{m1} & \cdots & a_{mn}
\end{bmatrix} = \begin{bmatrix}
    b_{1} \\
    \vdots \\
    b_{m}
\end{bmatrix}$
